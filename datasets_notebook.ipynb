```python
from datasets import GMM, Normal
from schedulers import NoiseScheduler
import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt
from jax.scipy.stats import norm
```

GMM model visualisation

```python
noise_scheduler = NoiseScheduler()
data_model = GMM(noise_scheduler)

key = jax.random.PRNGKey(0)
samples = data_model.generate_data(key, 10000)
```

```python
print(samples.shape)
x_values = jnp.linspace(-4, 4, 1000)
gt_densities = jnp.squeeze(data_model.density(jnp.expand_dims(x_values, axis=-1)))
less_noisy_densities = jnp.squeeze(data_model.noisy_density(jnp.expand_dims(x_values, axis=-1), 0.01))
more_noisy_densities = jnp.squeeze(data_model.noisy_density(jnp.expand_dims(x_values, axis=-1), 0.5))
plt.hist(jnp.squeeze(samples), density=True, bins=32)
plt.plot(x_values, gt_densities, label="gt density")
plt.plot(x_values, less_noisy_densities, label="less noisy density")
plt.plot(x_values, more_noisy_densities, label="more noisy density")
plt.plot(x_values, norm.pdf(x_values), label="standard normal")
plt.legend()
plt.show()
```

```python
x_values = jnp.linspace(-4, 4, 1000)
gt_score = jnp.squeeze(data_model.score(jnp.expand_dims(x_values, axis=-1)))
less_noisy_score = jnp.squeeze(data_model.noisy_score(jnp.expand_dims(x_values, axis=-1), 0.01))
more_noisy_score = jnp.squeeze(data_model.noisy_score(jnp.expand_dims(x_values, axis=-1), 0.4))
plt.plot(x_values, gt_score, label="gt score")
plt.plot(x_values, less_noisy_score, label="less noisy score")
plt.plot(x_values, more_noisy_score, label="more noisy score")
plt.plot(x_values, -x_values, label="standard normal")
plt.legend()
plt.show()
```
